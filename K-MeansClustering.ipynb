# K-Means Clustering

```python
# Import Pandas library to load and work with data.
import pandas as pd

# Import K-Means algorithm from scikit-learn.
from sklearn.cluster import KMeans

# Import Matplotlib to create graphs and visualize clusters.
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a Pandas DataFrame.
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head(2) 
```

```python
# Import Matplotlib library for plotting graphs.
import matplotlib.pyplot as plt

# Import K-Means algorithm from scikit-learn.
from sklearn.cluster import KMeans

# Select only the required columns from the dataset.
x = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Create an empty list to store WCSS values.
wcss = []

# Run K-Means for cluster values from 1 to 10.
for i in range(1, 11):

    # Create a K-Means model with i clusters.
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)

    # Train the model using the selected data.
    kmeans.fit(x)

    # Store the WCSS (Inertia) value in the list.
    wcss.append(kmeans.inertia_)

# Create a figure with width 10 and height 6.
plt.figure(figsize=(10, 6))

# Plot the Elbow Method graph.
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')

# Label the X-axis.
plt.xlabel('Number of Clusters')

# Label the Y-axis.
plt.ylabel('WCSS')

# Give a title to the graph.
plt.title('Elbow Method')

# Display the graph.
plt.show()
```

```python
# Create a K-Means model with 5 clusters.
kmeans = KMeans(n_clusters=5)

# Train the model and assign each data point to a cluster.
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
# Import Matplotlib library for plotting graphs.
import matplotlib.pyplot as plt

# Define different colors for each cluster.
colors = ['red', 'blue', 'pink', 'yellow', 'green']

# Loop through all 5 clusters.
for i in range(5):

    # Plot the data points of each cluster.
    plt.scatter(
        x.iloc[y_kmeans == i, 0],   # X-axis: Annual Income
        x.iloc[y_kmeans == i, 1],   # Y-axis: Spending Score
        s=60,                       # Size of points
        c=colors[i],                # Color of the cluster
        label=f'Cluster {i+1}'      # Cluster label
    )

# Plot the centroids (center points of clusters).
plt.scatter(
    kmeans.cluster_centers_[:, 0],  # X-coordinate of centroids
    kmeans.cluster_centers_[:, 1],  # Y-coordinate of centroids
    s=100,                          # Size of centroid
    c='black',                      # Color of centroid
    marker='o',                     # Shape of centroid
    label='Centroids'               # Label for legend
)

# Set the graph title.
plt.title('Clusters of Customers')

# Label the X-axis.
plt.xlabel('Annual Income (k$)')

# Label the Y-axis.
plt.ylabel('Spending Score (1-100)')

# Show the legend.
plt.legend()

# Display the graph.
plt.show()
```
