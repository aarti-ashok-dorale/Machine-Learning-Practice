## using classification(0/1)

```python
import pandas as pd

df = pd.DataFrame({
    'attendance': [90, 85, 70, 95, 60],
    'assignment': [88, 75, 65, 92, 58],
    'exam': [85, 72, 60, 95, 55],
    'success': [1, 1, 0, 1, 0]
})

df
```

```python
x = df[['attendance', 'assignment', 'exam']]
y = df['success']
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x,y) 
```

```python
new_data = pd.DataFrame({
    'attendance': [60],
    'assignment': [50],
    'exam': [55],
})

model.predict(new_data) 
```

## using regression

## predict car price based on car age 

```python
import pandas as pd
data={
    'car age' : [1,2,3,4,5,6,7],
    'car price(lakhs)' : [12,11,9.5,8,7,6,5]
}

car = pd.DataFrame(data)
car.head(2) 
```

```python
x = car[['car age']]
y = car['car price(lakhs)'] 
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors = 1)
model.fit(x_train,y_train) 
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
import numpy as np
car_age = np.array([[2.5]])
model.predict(car_age)
```
