# Decision Tree

## Eg; 1 on sample data

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# sample datase
data = {
    'Hours':[1,2,3,4,5,6],
    'Attendance': [0,0,1,1,1,1], #0=low, 1=high
    'Result': [0,0,0,1,1,1]     #0=fail, 1=pass
} 
```

```python
df = pd.DataFrame(data) 
x = df[['Hours', 'Attendance']]
y = df['Result']
```

```python
#train, test split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
model =  DecisionTreeClassifier(criterion="entropy", max_depth=3)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)  
```

```python
from sklearn.tree import plot_tree 
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))
plot_tree(model, feature_names=['Hours', 'Attendance'], class_names=['Fail', 'Pass'], filled=True)
plt.show()
```

# maths behind code 

## entropy

```python
# maths
import math
def entropy(labels):
    total = len(labels) 

    count1 = labels.count('yes')
    count2 = labels.count('no')

    p1 = count1/total
    p2 = count2/total
    
    ent = 0
    if p1>0:
        ent=-p1*math.log2(p1)
    
    if p2>0:
        ent=-p2*math.log2(p2)

    return ent

labels = ['yes','yes','yes','yes','yes','yes','no','no','no','no']
entropy(labels)   
```

## gini index

```python
# gini index
def gini(labels):
    total=len(labels)

    count1 = labels.count('yes')
    count2 = labels.count('no')

    p1 = count1/total
    p2 = count2/total

    return 1-(p1**2 + p2**2)

labels = ['yes','yes','yes','yes','yes','yes','no','no','no','no']
gini(labels) 
```

## IG = Information Gain 

```python
parent_entropy = 0.97
left_entropy = 0.50
right_entropy = 0.20
left_size = 6 
right_size = 4

total = left_size + right_size
weighted_entropy = ((left_size/total)*left_entropy+(right_size/total)*right_entropy)

IG = parent_entropy - weighted_entropy 
print(IG)
```

# DecisionTreeRegressor

# 1.Regressor : diabetes dataset (CART)

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
```

```python
data  = load_diabetes()
x = data.data
y = data.target
data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
model = DecisionTreeRegressor(random_state = 42, max_depth = 3)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
plt.figure(figsize = (15,8))
plot_tree(model,feature_names = data.feature_names, filled = True, rounded = True, fontsize = 10)
plt.title("DecisionTreeRegressor")
plt.show()
```

# 2. Regressor : Housing dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing
```

```python
house = fetch_california_housing()
```

```python
x = house.data
y = house.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
model = DecisionTreeRegressor(max_depth = 4)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize = (30,10))
plot_tree(model,feature_names = house.feature_names, filled = True, fontsize = 18) 
```
