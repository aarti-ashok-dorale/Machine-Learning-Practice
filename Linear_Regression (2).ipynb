```python
# step1.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt 
```

```python
# step2. collect data

data = {
    'size(sqft)' :[1000, 2000, 3000, 4000, 5000,],
    'price(lpa)' :[20, 40, 60, 80, 100]
}

df = pd.DataFrame(data)
df
```

```python
# step 3. select feature and target values

x = df[["size(sqft)"]]
y = df['price(lpa)']
```

```python
plt.scatter(x,y)  # scatters/bubbles display
plt.plot(x,y) # line display
plt.xlabel("size(sqft)") # x axis title
plt.ylabel('size(lpa)') # y axis title
plt.title("Linear Regression") # graph title
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#apply model

model = LinearRegression()

#train model

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
#evaluate model

r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred) 
```

```python
#new data prediction

new_data = pd.DataFrame(
    {
        "size(sqft)" : [1500, 1800, 2400]
    }
)

model.predict(new_data)
```
